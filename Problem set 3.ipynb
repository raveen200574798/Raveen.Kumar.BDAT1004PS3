Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python

def fix_century(x):
  year = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(year, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)

data.head()
```

# <b>Question 8</b>

Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name.

Don't forget to label your axes!

For reference, here's the raw data:

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = r"C:\Users\ravee\Downloads\actor_kill_counts.csv"
hollywood_actor_kills = pd.read_csv(file_path)

sorted_actors = hollywood_actor_kills.sort_values(by='Count', ascending=True)

actor_names = sorted_actors['Actor'].values
kill_counts = sorted_actors['Count'].values

plt.figure(figsize=(10, 8))
plt.barh(actor_names, kill_counts, color='skyblue')
plt.xlabel('Kill Counts')
plt.ylabel('Actors')
plt.title('Deadliest Actors in Hollywood by Kill Count')
plt.tight_layout()

for index, value in enumerate(kill_counts):
    plt.text(value, index, str(value))

plt.show()
```

#  <b>Question 9</b>

Create a pie chart showing the fraction of all Roman Emperors that were assassinated.

Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\ravee\Downloads\roman-emperor-reigns.csv"
roman_emperors = pd.read_csv(file_path)

assassinated_emperors = roman_emperors[
    roman_emperors['Cause_of_Death'].str.lower().str.contains('assassinated')]

number_assassinated = len(assassinated_emperors)
other_deaths = len(roman_emperors) - number_assassinated

labels = ['Assassinated Emperors', 'Other Deaths']
sizes = [number_assassinated, other_deaths]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal') 
plt.title('Fraction of Roman Emperors Assassinated')
plt.show()
```

Create a scatter plot showing the relationship between the total revenue earned byarcades and the number of Computer Science PhDs awarded in the U.S. between2000 and 2009.

Don't forget to label your axes!

Color each dot according to its year

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\ravee\Downloads\arcade-revenue-vs-cs-doctorates.csv"
arcade_revenue_cs_doctorates = pd.read_csv(file_path)

arcade_revenue = arcade_revenue_cs_doctorates['Total Arcade Revenue (billions)'].values
cs_doctorates_awarded = arcade_revenue_cs_doctorates['Computer Science Doctorates Awarded (US)'].values
years = arcade_revenue_cs_doctorates['Year'].values

plt.figure(figsize=(10, 6))
for i, year in enumerate(years):
    plt.scatter(arcade_revenue[i], cs_doctorates_awarded[i], label=str(year))

plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.title('Relationship between Arcade Revenue and CS Doctorates Awarded (2000-2009)')
plt.legend(title='Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```
